```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Topic_Survey_Assignment.csv')
df
```

```python
df.sort_values(['Very interested'],ascending=False, axis= 0, inplace=True)
df
```

```python
respondents = 2233
df['Very interested'] = df['Very interested']/ respondents * 100
df['Somewhat interested'] = df['Somewhat interested']/ respondents * 100
df['Not interested'] = df['Not interested']/ respondents * 100
df
```

```python
df.round(2)
```

```python
    per = df.plot(kind='bar', 
                figsize=(20, 8), 
                width=0.8,
                color=['#5cb85c', '#5bc0de', '#d9534f'],
               )
    per.set_title("Percentage of Respondents' Interest in Data Science Area", fontsize=16),
    per.spines['left'].set_visible(False),
    per.spines['top'].set_visible(False),
    per.spines['right'].set_visible(False),
    per.axes.get_yaxis().set_visible(False),
    per.tick_params(labelsize=14),
    per.legend(fontsize=14),
    for p in per.patches:
        per.annotate(np.round(p.get_height(),decimals=2),
                    (p.get_x()+p.get_width()/2., p.get_height()), 
                    ha='center',
                    va='center', 
                    xytext=(0, 10),
                    textcoords='offset points',
                    fontsize = 14,
                   ),
    labels = list[df.columns]
    plt.xlabel(labels)    
    plt.show()
   
```
